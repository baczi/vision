## Create an Resource Group

As its name suggests, a workspace is a centralized place to manage all of the Azure ML assets you need to work on a machine learning project.

1. In the [Azure portal](https://portal.azure.com), create a new **Resource Group**, using CLI :




Open CloudShell in Azure Portal.

Select **bash** in drop-down menu 



In code past next command: 

```python
az group create --name demoOCR --location westeurope
```

This command will create us a new Resource Group. Wait for fully deployment. When it's finished deploy, We will create all needed resources.

### Create resource - ComputerViision 

```python
az cognitiveservices account create --name demoCV  --resource-group demoOCR  --kind ComputerVision --sku F0 --location westeurope  --yes
```

### Create storage account as a datalake

```python
az storage account create --resource-group demoOCR --name <your-unique-name> --location westeurope --sku Standard_LRS --kind StorageV2 --enable-hierarchical-namespace true
```
